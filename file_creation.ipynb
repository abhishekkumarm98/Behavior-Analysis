```python
def getResiduals(file,file1,reward):
    file =file[file['ID'].notnull()].reset_index(drop=True)
    residuals = []
    expected_ip =[]
    impulsitivity = []
    for idx in range(file.shape[0]):
        try:
            k=file1[file1['id']==file.iloc[idx]['ID']].values[0][2] 
            impulsitivity.append(k)
            pred_ip = reward /(1+file.iloc[idx]['DELAY']*k)
            expected_ip.append(pred_ip)
            if reward == 100:
                resd = file.iloc[idx]['IP VALUE'] - pred_ip 
                residuals.append(resd)
                
            elif reward == 1000:
                resd = file.iloc[idx]['IP_1000_VALUE'] - pred_ip
                residuals.append(resd)
                
                
        except IndexError:
            impulsitivity.append(np.nan)
            expected_ip.append(np.nan)
            residuals.append(np.nan)
    
    file['Expected_IP'] = expected_ip
    file['Impulsitivity Effect/ Residuals'] = residuals
    file['Impulsitivity/K'] = impulsitivity
    return file
```
